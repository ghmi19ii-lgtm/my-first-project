Phase 2: Baseline CNN Model (Initial Training)

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
import kagglehub
import os
```

```python
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
```

```python
path = kagglehub.dataset_download("abdallahalidev/plantvillage-dataset")
```

```python
dataset_path = os.path.join(path, 'plantvillage dataset', 'segmented')
print("Ш§Щ„Щ…ШіШ§Ш± Ш§Щ„Щ…Ш№ШЄЩ…ШЇ Щ‡Щ€:", dataset_path)
```

```python
train_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='training')

validation_generator = datagen.flow_from_directory(
    dataset_path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation')
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(train_generator.num_classes, activation='softmax') # ШіЩЉШЈШ®Ш° Ш§Щ„Ш№ШЇШЇ Ш§Щ„ШµШ­ЩЉШ­ ШЄЩ„Щ‚Ш§Ш¦ЩЉШ§Щ‹
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Ш§Щ„ШЄШЇШ±ЩЉШЁ
history = model.fit(train_generator, epochs=5, validation_data=validation_generator)
```

Phase 3: Optimized Model with Dropout & Augmentation

```python
# Image augmentation to improve model generalization
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2
)
```

```python
model = models.Sequential([
    # First Convolutional Block
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    layers.MaxPooling2D(2,2),

    # Second Convolutional Block
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    # Third Convolutional Block
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    # Flattening and Dense Layers
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5), # Prevents overfitting
    layers.Dense(train_generator.num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Training for 10 epochs to reach higher stability
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=validation_generator
)
```
